```python
df1 = pd.read_csv("study4_trials.csv")
df2 = pd.read_csv("study4_answers.csv")
```

Vi läser in de två nya csv-filerna

```python
#Omvandlar datan med trials till lång data

df_trials_long = df1.melt(
    id_vars='PID',
    var_name='trial',
    value_name='choice'
)

df_trials_long
```

```python
#Omvandlar datan med svaren på uppgiften till lång data
df_answers_long = df2.melt(
    id_vars='PID',
    var_name='trial',
    value_name='answer'
)

df_answers_long
```

```python
#Dubbelcheckar alla unika värden i kolumnen "answer"
df_answers_long['answer'].unique()
```

```python
#Rättar till
df_answers_long.replace({ 
'answer': {'incorrect0': 1, 'blank': 0}
}, inplace=True)
```

```python
df_answers_long['answer'].unique()
```

**OBS:** Just nu är vi osäkra på varför vissa siffror har citattecken, så låter dem vara just nu. 

```python
#Ta bort 'ans' på slutet av strängarna i kolumnen trial 
df_answers_long['trial'] = df_answers_long['trial'].str.replace('ans', '', regex=False)
```

För att kunna merge:a de två csv-filerna behöver de ha matchande kolumner, så vi tar bort "ans" på värderna i kolumnen 'trial' i answers-filen. Nu har vi merge:a med PID och trial. (regex=False ser till att bara ans tas bort)

```python
df_merge = pd.merge(df_trials_long, df_answers_long, on =['PID', 'trial'], how='inner')

df_merge
```

Nu har vi merge:at filerna. Vi har nu en tabell med fyra kolumner: **'PID'**: deltagare, **'trial'**: försök 1-40, **'choice'**: Gör uppgiften(2)/Gör inget(1), **'answer'**: om choice = 2 answer: 2(korrekt svar)/1(fel svar), om choice = 1 answer: 0
